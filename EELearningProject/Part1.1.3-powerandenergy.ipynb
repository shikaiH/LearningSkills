## Power Theorem

Power is the time rate of expanding or absorbing energy, measured in watts(W).


p is power in watts(W), w is charge in joules(J), t is time in seconds(s).

$p \triangleq \frac{dw}{dt}$

$p = \frac{dw}{dt} = \frac{dw}{dq} \cdot \frac{dq}{dt} = vi$

or 

$p = vi = watts$

voltage is measured in volts, i is current in amps.
