## What is voltage?

The electron in a conductor in a particular directions requires some work or energy transfer.

This work is performed by an external electromative force(emf), also known as voltage or potential difference.

The voltage between two points a and b in an electric circuit is the work needed to move a unit charge from b to a.

## Voltage Theorem

Voltage or potential difference is the engery required to move a unit charge from a reference point (-) to another point (+), measured in volts(V).



w is energy in joule, q is charge in coulombs, v is measured in volts.

$v_{ab} \triangleq \frac{dw}{dq}$

$1 \text{ volt} = 1 \text{ joule/coulomb} = 1 \text{ newton-meter/coulomb}$

$v_{ab} = -v_{ba}$

voltage drop from a to b is equivalent to a voltage rise from b to a.

V -> dc voltage constant voltage (capital V)

v -> ac voltage varing with the time (small v)
